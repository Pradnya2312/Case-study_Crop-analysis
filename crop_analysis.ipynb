```python
plt.boxplot(df['Production'])
plt.show()
```

```python
max_production = df['Production'].max()
max_production_crop = df[df['Production'] == max_production]['Crop'].values[0]

print("Maximum Production:", max_production)
print("Crop with Maximum Production:", max_production_crop)
```

```python
state_crop_year = df.groupby(['State_Name', 'Crop_Year', 'Crop'])['Production'].sum().reset_index()

print("State-wise Crop Production and Their Years:")
print(state_crop_year)
```

```python
# Group by State to find the total production for each state
state_production = df.groupby('State_Name')['Production'].sum()

# Find the state with maximum and minimum production
max_production_state = state_production.idxmax()
min_production_state = state_production.idxmin()

print("State with Maximum Production:", max_production_state)
print("State with Minimum Production:", min_production_state)
```

```python

# Group by Season and Crop to find the total production for each crop in each season
season_crop_production = df.groupby(['Season', 'Crop'])['Production'].sum().reset_index()

# Find the maximum production crop for each season
max_production_per_season = season_crop_production.loc[season_crop_production.groupby('Season')['Production'].idxmax()]

print("Season-wise Maximum Crop Production:")
print(max_production_per_season)

# Create a bar plot for the season-wise maximum crop production
plt.figure(figsize=(10, 5))
plt.bar(max_production_per_season['Season'], max_production_per_season['Production'], color='blue', alpha=0.7)

# Add title and labels
plt.title('Season-wise Maximum Crop Production')
plt.xlabel('Season')
plt.ylabel('Production')

# Annotate the bars with crop names
for i, crop in enumerate(max_production_per_season['Crop']):
    plt.text(i, max_production_per_season['Production'].iloc[i] + 500, crop, ha='center', va='bottom')

# Display the plot
plt.show()
```

```python
# Group by Season and Crop to find the total production for each crop in each season
season_crop_production = df.groupby(['Season', 'Crop'])['Production'].sum().reset_index()

# Find the maximum production crop for each season
min_production_per_season = season_crop_production.loc[season_crop_production.groupby('Season')['Production'].idxmin()]

print("Season-wise Minimum Crop Production:")
print(min_production_per_season)
```
